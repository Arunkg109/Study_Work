```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

data = pd.read_csv('glass1.data')
data.head()
```

## How to reduce dimention

```python
data.isna().sum()   #checking for null values
```

```python
data = data.drop(labels=['index','Class'],axis = 1)
```

```python
data  #file chage
```

```python
# Convert data into PCA

#1 Transforming our dataset

from sklearn.preprocessing import StandardScaler
```

```python
#2 Class attribute..
scaler = StandardScaler()  # it'll act as blueprint
```

```python
scaled_data = scaler.fit_transform(data)
```

```python
scaled_data
```

```python
# change vector otation to dataframe

df = pd.DataFrame(data = scaled_data, columns=data.columns)
df
```

```python
# load PCA (Principle component analysis) as PCA class
from sklearn.decomposition import PCA
```

```python
# PCA instance
pca = PCA()
```

```python
pr_components = pca.fit_transform(df)
pr_components
```

```python
 # plot a graph... EVR
    
    
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('No. Of components')
plt.ylabel('Variece %')
plt.title('Explained variance ration')
```

```python
pca = PCA(n_components=5)
pca
```

```python
# only 5 columns..
new_data = pca.fit_transform(df)
new_data   # 
```

```python
# convert new_data in dataframe

Final_data = pd.DataFrame(data = new_data, columns=['PC1', 'PC2','PC3','PC4','PC5'])
Final_data
```
